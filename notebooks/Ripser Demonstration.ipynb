# Ripser demonstration

This notebook shows the most basic use of the Ripser Python API.

```python
%load_ext autoreload
%autoreload 2
```

```python
from ripser import Rips
import numpy as np
from sklearn import datasets
```

```python
data = datasets.make_circles(n_samples=100)[0] + 5 * datasets.make_circles(n_samples=100)[0]
```

# Default args make it very easy. 

- Generate diagrams for $H_0$ and $H_1$
- Plot both diagrams

```python
rips = Rips()
dgms = rips.fit_transform(data)
rips.plot()
```

```python
# Get each diagram by plotted by itself
H_0, H_1 = dgms
rips.plot(H_0)
rips.plot(H_1)
```

# Homology over another basis

We can compute homology over any $p\ge 2$ by supplying the argument `coeff=p`.

```python
# Homology over Z_3
rips = Rips(coeff=3)
dgms = rips.fit_transform(data)
rips.plot(dgms[1]) # Only plot H_1
```

# Specify which homology

We can compute any order of homology, $H_0, H_1, H_2, \ldots$. By default, we only compute $H_0$ and $H_1$. You can specify a larger by supplying the argument `maxdim=p`. It practice, anything above $H_1$ is very slow.

```python
rips = Rips(maxdim=2)
dgms = rips.fit_transform(data)
print(len(dgms))
rips.plot() 
```

# Specify maximum radius for Rips filtration

We can restrict the maximum radius of the VR complex by supplying the argument `thresh=r`.

```python
rips = Rips(thresh=0.2)
dgms = rips.fit_transform(data)
rips.plot()
```

```python
rips = Rips(thresh=1)
dgms = rips.fit_transform(data)
rips.plot()
```

```python
rips = Rips(thresh=999)
dgms = rips.fit_transform(data)
rips.plot()
```
